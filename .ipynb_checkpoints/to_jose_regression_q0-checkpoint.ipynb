```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt

class Q1(tf.keras.Model):
    def __init__(self,output_dim=1):
        super(Q1,self).__init__()

        self.l1 = Dense(33, input_shape=(1,))
        self.l2 = Dense(33)
        self.l3 = Dense(output_dim)

    def call(self, input):
        feat = tf.nn.sigmoid(self.l1(input))
        feat = tf.nn.sigmoid(self.l2(feat))
        value = tf.nn.sigmoid(self.l3(feat))
        return value

    def give_first_beta(self, epsilon):
        if np.random.random() < epsilon:
            label = np.random.choice(np.arange(len(self.betas)))
            return label, self.betas[label]
        else:
            input = np.expand_dims(np.array([]), axis=0)
            q1s = self.call(input)
            q1s = q1s.numpy()
            label = np.argmax(q1s)
            beta1 = self.betas[label]
            return label, beta1

```

```python
betis = np.arange(.5,1,.1)
```

```python
def Prob(alpha, beta, n1):
    p0 = np.exp(-(alpha-beta)**2)
    if n1 == 0:
        return p0
    else:
        return 1-p0
    
def ps(beta):
    alpha = 0.4
    p=0
    for n1 in [0,1]:
       p+=np.max([Prob(ph*alpha, beta, n1) for ph in [-1,1]]) 
    return p/2
```

```python
q1 = Q1()

optimizer = tf.keras.optimizers.Adam(lr = 10E-3)

betas_batch = np.expand_dims(np.array( [i for i in betis] ), axis=1)
labels_batch = np.expand_dims(np.array( [ps(i) for i in betis] ), axis=1)
betas_test = np.arange(.5,1,.01)
betas_test = np.expand_dims(np.array( [i for i in betas_test] ), axis=1)
```

```python

loss_ev = []
total=10**4
times_saving = []
preds = []
for k in range(int(np.log10(total))):
    for t in np.arange(10**k,10**(k+1)+1,10**k):
        times_saving.append(t)
for episode in range(total):
    if episode%int(total/10) == 0:
        print(episode,"/",total)
    with tf.device("/cpu:0"):
        with tf.GradientTape() as tape:
            tape.watch(q1.trainable_variables)
            predictions = q1(betas_batch)
            loss_sum = tf.keras.losses.MSE(predictions,labels_batch)
            loss = tf.reduce_mean(loss_sum)
            grads = tape.gradient(loss, q1.trainable_variables)
            optimizer.apply_gradients(zip(grads, q1.trainable_variables))
    loss_ev.append(loss.numpy())
    preds.append(np.squeeze(q1(betas_test).numpy()))
```

```python
def show_results():
    plt.figure(figsize=(20,20),dpi=40)
    plt.rcParams.update({'font.size': 50})

    ax1 = plt.subplot2grid((2,1), (0,0))
    ax2 = plt.subplot2grid((2,1),(1,0))
    betas_test = np.expand_dims(np.array( [i for i in np.arange(.5,1,.01)] ), axis=1)

    ax1.plot(np.log10(np.arange(1,len(loss_ev)+1)),loss_ev, linewidth=8,label="Loss")
    ax2.plot(np.arange(.5,1,.01),np.squeeze(q1(betas_test).numpy()), '.',linewidth= 8 ,color="red",label="pred")
    ax2.scatter(betis,np.squeeze(labels_batch),color="blue",  s=130, label="true value")
    ax2.plot(np.arange(.5,1,.01),[ps(b) for b in np.arange(.5,1,.01)],'--', color="black", label="P_s")

    for ax in [ax1,ax2]:
        ax.legend(prop={"size":40})
show_results()
```
